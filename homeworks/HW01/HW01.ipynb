```python
a = np.random.randint(1,100, (2, 2))
print(a)
```

```python
c = np.hstack((a,b)) # Горизонтальное объединение массивов
print(c)
```

```python
np.mean(c) # Среднее значение
```

```python
np.hsplit(c,(1,2)) # Разрезать c после 1 и 2 столбца
```
